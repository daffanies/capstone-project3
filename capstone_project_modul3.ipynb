## **Business Problem Understanding**

**Context**

Apartemen merupakan salah satu solusi kebutuhan perumahan masyarakat modern, mengingat keterbatasan lahan dan tingginya aktivitas bisnis di perkotaan. Oleh karena itu, akan sangat menarik untuk menganalisis harga apartemen yang dipengaruhi oleh berbagai faktor internal dan eksternal.

Individu atau perusahaan biasanya menawarkan unit apartemen. Penjual dapat memasarkan unit mereka melalui platform tertentu dengan menentukan harga apartemen mereka sendiri. Cara ini cukup sulit bagi pemilik apartemen untuk menyesuaikan dengan harga pasar. Jika harga terlalu tinggi dibandingkan harga pasar, tentu akan sulit untuk menjual unit tersebut. Sebaliknya, jika terlalu rendah, pemilik akan kesulitan mendapatkan keuntungan maksimal.

**Problem Statement**

Salah satu tantangan utama dalam industri properti, khususnya di segmen apartemen, adalah ketidakpastian dalam menentukan harga jual yang optimal. Fluktuasi pasar yang dinamis, serta beragamnya faktor internal dan eksternal yang mempengaruhi harga properti, membuat pemilik apartemen kesulitan untuk menentukan harga jual yang kompetitif dan realistis. Akibatnya, apartemen mungkin dijual dengan harga yang terlalu tinggi sehingga sulit laku, atau sebaliknya, dijual terlalu murah sehingga pemilik kehilangan potensi keuntungan. 

Untuk mengatasi permasalahan ini, diperlukan sebuah model prediksi harga yang akurat dan komprehensif, yang dapat mempertimbangkan berbagai faktor seperti lokasi, luas bangunan, fasilitas serta kondisi pasar saat ini. **Dengan adanya model prediksi harga yang handal, pemilik apartemen dapat membuat keputusan yang lebih informatif dan efektif dalam menentukan harga jual properti mereka.**

**Goals**

Tujuan utama dari pengembangan model prediksi harga apartemen ini adalah **untuk memberikan alat bantu yang akurat dan efisien bagi pemilik apartemen dalam menentukan harga jual yang optimal.** Dengan memanfaatkan data historis harga properti, karakteristik fisik apartemen seperti luas, fasilitas, dan lokasi, serta faktor eksternal seperti aksesibilitas transportasi publik dan keberadaan fasilitas umum di sekitar, model ini diharapkan dapat menghasilkan prediksi harga jual yang lebih realistis dan kompetitif. 

Melalui model ini, pemilik apartemen dapat memaksimalkan keuntungan penjualan, menghindari kerugian akibat harga yang tidak sesuai pasar, dan mempercepat proses transaksi. Selain itu, model ini juga dapat memberikan insights yang berharga bagi pengembang properti, investor, dan pembuat kebijakan dalam memahami dinamika pasar properti serta mengidentifikasi peluang investasi yang potensial.

**Analytic Approach**

Untuk membangun model prediksi harga jual apartemen yang akurat, kita akan melakukan analisis data secara mendalam. 

Langkah pertama adalah mengidentifikasi korelasi antara fitur-fitur numerik seperti luas apartemen (Size), tahun pembangunan (YearBuilt), dan jumlah fasilitas (N_FacilitiesInApt) dengan harga jual (SalePrice). Selain itu, kita juga akan menganalisis pengaruh faktor kategorikal seperti tipe apartemen (Hallway Type) dan keberadaan fasilitas umum di sekitar (N_FacilitiesNearBy) terhadap harga. Dengan menggunakan teknik-teknik statistik seperti korelasi dan regresi linier berganda, kita dapat mengidentifikasi pola-pola hubungan yang signifikan antara variabel-variabel tersebut. 

Selanjutnya, model regresi yang dihasilkan akan digunakan untuk memprediksi harga jual apartemen baru berdasarkan nilai fitur-fiturnya. Model ini diharapkan dapat memberikan rekomendasi harga yang akurat dan dapat diandalkan bagi pemilik apartemen.

**Metric Evaluation**

Untuk mengevaluasi performa model prediksi harga jual apartemen, kita akan menggunakan beberapa metrik evaluasi. Metrik yang umum digunakan dalam masalah regresi adalah Root Mean Squared Error (RMSE), Mean Absolute Error (MAE), dan Mean Absolute Percentage Error (MAPE). RMSE memberikan gambaran tentang seberapa besar kesalahan prediksi secara rata-rata, sedangkan MAE memberikan nilai absolut dari kesalahan prediksi. MAPE, di sisi lain, memberikan persentase kesalahan prediksi terhadap nilai sebenarnya. Semakin kecil nilai ketiga metrik tersebut, semakin baik kinerja model dalam memprediksi harga jual.

Selain itu, kita juga akan menghitung nilai R-squared untuk model-model linear. Nilai R-squared menunjukkan proporsi variabilitas dalam variabel dependen (harga jual) yang dapat dijelaskan oleh variabel independen (fitur-fitur apartemen). Nilai R-squared yang mendekati 1 mengindikasikan bahwa model dapat menjelaskan sebagian besar variasi dalam data.

### Outliers

```python
# Outliers Boxplot

plt.figure(figsize=(12,10))
sns.boxplot(data=df_apt)
plt.title('Boxplot')
plt.xticks(rotation=90)
plt.show()
```

```python
# Fungsi untuk mendeteksi outlier menggunakan metode IQR

def deteksi_outlier(df, kolom):
    Q1 = df[kolom].quantile(0.25)
    Q3 = df[kolom].quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR
    outlier = df[(df[kolom] < batas_bawah) | (df[kolom] > batas_atas)]
    return outlier

# Mendapatkan daftar kolom numerik

kolom_numerik = df_apt.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
# Menghitung jumlah outlier pada setiap kolom

jumlah_outlier = {}
for kolom in kolom_numerik:
    outlier = deteksi_outlier(df_apt, kolom)
    jumlah_outlier[kolom] = len(outlier)

print("\nJumlah outlier pada setiap kolom numerik:")
for kolom, jumlah in jumlah_outlier.items():
    print(f"- {kolom}: {jumlah} outlier")
```

```python
# Menghitung persentase outlier pada setiap kolom

persentase_outlier = {}
for kolom, jumlah in jumlah_outlier.items():
    total_data = len(df_apt)
    persentase = (jumlah / total_data) * 100
    persentase_outlier[kolom] = persentase

print("\nPersentase outlier pada setiap kolom numerik:")
for kolom, persentase in persentase_outlier.items():
    print(f"- {kolom}: {persentase:.2f}%")
```

Variabel seperti jumlah fasilitas terdekat, jumlah tempat parkir basement, tahun pembangunan, dan jumlah fasilitas dalam apartemen memiliki distribusi data yang relatif bersih tanpa adanya nilai ekstrem yang signifikan. 

Namun, variabel `Size(sqf)` dan `SalePrice` ditemukan memiliki sejumlah outlier, masing-masing sebesar `3.11%` dan `0.63%` dari total data. Hal ini mengindikasikan adanya beberapa properti dengan ukuran atau harga yang jauh berbeda dari rata-rata, yang mungkin memerlukan analisis lebih lanjut untuk memahami penyebab dan dampaknya terhadap model prediksi.

```python
# Fungsi untuk mengatasi outlier

def atasi_outlier(df, kolom):
    Q1 = df[kolom].quantile(0.25)
    Q3 = df[kolom].quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR
    
    df[kolom] = df[kolom].clip(lower=batas_bawah, upper=batas_atas)
    return df

# Mengatasi outlier pada setiap kolom numerik

for kolom in kolom_numerik:
    df_apt = atasi_outlier(df_apt, kolom)

# Memeriksa kembali jumlah outlier setelah penanganan

jumlah_outlier_setelah = {}
for kolom in kolom_numerik:
    outlier = deteksi_outlier(df_apt, kolom)
    jumlah_outlier_setelah[kolom] = len(outlier)

print("\nJumlah outlier setelah penanganan:")
for kolom, jumlah in jumlah_outlier_setelah.items():
    print(f"- {kolom}: {jumlah} outlier")
```

Hasil menunjukkan bahwa setelah dilakukan penanganan, tidak ditemukan lagi outliers pada semua kolom numerik yang diperiksa. Kolom-kolom tersebut mencakup berbagai fitur apartemen terutama pada Size(sqf) dan SalePrice

Fitur fitur yang ada dalam dataset akan di kelompokan menjadi tipe numerikal, kategorikal, dan ordinal. Langkah ini penting untuk mempersiapkan data sebelum masuk ke pipeline pemodelan. Pengelompokan ini memungkinkan penerapan encoding dan scaling yang tepat pada setiap jenis fitur. 

Dengan demikian, agar dapat memastikan bahwa data akan diolah secara akurat dan komprehensif, meningkatkan kualitas input untuk proses pemodelan. Hal ini membantu mengoptimalkan kinerja model dan meningkatkan interpretabilitas hasil.

```python
# Library untuk Feature Engineering

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from category_encoders import OrdinalEncoder as ce_OrdinalEncoder
```

```python
df_apt.info()
```

```python
# Mengelompokkan fitur berdasarkan tipe data

num_feat = ['N_FacilitiesNearBy(ETC)', 'N_FacilitiesNearBy(PublicOffice)', 'N_SchoolNearBy(University)',
            'N_Parkinglot(Basement)', 'YearBuilt', 'N_FacilitiesInApt', 'Size(sqf)']
cat_feat = ['HallwayType', 'SubwayStation']
ord_feat = ['TimeToSubway']

# Ordinaliasi / mengurutkan nilai pada fitur TimeToSubway
ord_TimeToSubway = [
    {
        'col': 'TimeToSubway',
        'mapping': {
            'no_bus_stop_nearby': 1,
            '15min-20min': 2,
            '10min-15min': 3,
            '5min-10min': 4,
            '0-5min': 5
        }
    }
]
```

```python
# Preprocessing: Encoding & Scalling

prepros = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_feat),
        ('cat', OneHotEncoder(drop='first'), cat_feat),
        ('ord', ce_OrdinalEncoder(mapping=ord_TimeToSubway), ord_feat)
    ],
    remainder='passthrough'
)

# Pipeline untuk Preprocessing

df_apt_pipe = Pipeline(steps=[
    ('preprocessor', prepros),
])

df_apt_pipe
```

```python
# Melakukan preprocessing pada data

df_preprocessed = pd.DataFrame(df_apt_pipe.fit_transform(df_apt))

# Mendapatkan nama kolom yang sesuai

num_cols = num_feat
cat_cols = [f"{col}_{val}" for col in cat_feat for val in df_apt[col].unique() if val != df_apt[col].unique()[0]]
ord_cols = ord_feat

# Menggabungkan semua nama kolom

all_cols = num_cols + cat_cols + ord_cols + ['SalePrice']

# Menetapkan nama kolom yang sesuai

df_preprocessed.columns = all_cols
df_preprocessed
```

Setelah dilakukan preprocessing, dataset menjadi 18 kolom dan semua tipe datanya berupa float64. Fitur numerik telah dinormalisasi, fitur kategorikal telah diubah, dan fitur ordinal telah diubah menjadi data numerik. Dataset siap digunakan untuk pemodelan.

### Benchmarking Model

```python
# Library untuk Modeling

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost.sklearn import XGBRegressor
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score, mean_squared_error
```

```python
# Algoritma yang digunakan

linreg = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
svr = SVR()
xgb = XGBRegressor(random_state=1)
```

```python
# Scaling data

scaler = StandardScaler()

# Kandidat algoritma

models = [linreg, knn, dt, rf, svr, xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

score_r2 = []
nilai_mean_r2 = []
nilai_std_r2 = []

# Mencari algoritma terbaik berdasarkan tiap metrics

for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    # Gunakan X dan y langsung, tanpa preprocessing tambahan

    estimator = Pipeline([
        ('scaler', scaler),
        ('model', i)
    ])

    # RMSE

    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
    )

    print(model_cv_rmse, i)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE

    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
    )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE

    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
    )

    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())

    # R2 SCORE
    
    model_cv_r2 = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='r2', 
        error_score='raise'
    )

    print(model_cv_r2, i)

    score_r2.append(model_cv_r2)
    nilai_mean_r2.append(model_cv_r2.mean())
    nilai_std_r2.append(model_cv_r2.std())
```

#### Hyperparameter Tuning

```python
# Definisikan fungsi cross validation
crossval = KFold(n_splits=5, shuffle=True, random_state=1)

# Definisikan scoring metrics
scoring = {
    'rmse': 'neg_root_mean_squared_error',
    'mae': 'neg_mean_absolute_error',
    'mape': 'neg_mean_absolute_percentage_error',
    'r2': 'r2'
}
```

**`XGBoost`**

```python
# Parameter grid untuk XGBoost

param_grid_xgb = {
    'model__n_estimators': [100, 300],
    'model__max_depth': [3, 7],
    'model__learning_rate': [0.01, 0.1],
    'model__subsample': [0.8, 1.0],
    'model__colsample_bytree': [0.8, 1.0],
    'model__min_child_weight': [1, 5],
    'model__gamma': [0, 0.2],
    'model__reg_alpha': [0, 1],
    'model__reg_lambda': [0, 1]
}

# Pipeline XGBoost

xgb_pipeline = Pipeline([
    ('scaler', scaler),
    ('model', XGBRegressor(random_state=1))
])
```

```python
# Lakukan grid search dengan cross validation

grid_search_xgb = GridSearchCV(xgb_pipeline, param_grid_xgb, cv=crossval, scoring=scoring, refit='rmse', n_jobs=-1)
grid_search_xgb.fit(X_train, y_train)
```
